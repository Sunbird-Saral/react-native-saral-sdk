```python
path = '/home/venkateshiyer/Downloads/by_class/39/hsf_0/*'
result = {}
images = []
for image in sorted(glob.iglob(path)):
    img2 = cv2.imread(image,0)
    image_class = image.split('/')[-2]
    image_name = image.split('/')[-1]
    ret, thresh2 = cv2.threshold(img2, 127, 255,0)
    contours2,hierarchy = cv2.findContours(thresh2,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    cnt2 = contours2[-1]
    ret12 = cv2.matchShapes(cnt1,cnt2,1,0.0)
    if ret12 < 1.15:
        print(image_name, ret12)
        images.append(image_name)
        cv2.imwrite('/home/venkateshiyer/matching_digits/9/'+image_name, img2)
result[image_class] = images
```
